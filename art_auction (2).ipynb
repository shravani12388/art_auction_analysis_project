```python
# Average price by style
avg_price_style = art.groupby("style")["price_usd"].mean().sort_values(ascending=False)
print("\nAverage Price by Art Style:")
print(avg_price_style)
```

```python
# ðŸ“Š Count of artworks per artist per country
artist_country_counts = art.groupby(["country", "artist"]).size().sort_values(ascending=False).head(10)
print("\nTop 10 Country-Artist Artwork Counts:")
print(artist_country_counts)
```

```python
# Top 10 artists by number of artworks
top_artists = art['artist'].value_counts().head(10)
top_artists.plot(kind='bar', color='orchid')
plt.title("Top 10 Artists by Number of Artworks")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
# Price distribution
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(art["price_usd"], bins=50, color='skyblue')
plt.title("Price Distribution")
plt.xlabel("Price (USD)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Average price by style
avg_price_style = art.groupby('style')["price_usd"].mean().sort_values(ascending=False)

avg_price_style.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average Price by Style")
plt.xlabel("Style")
plt.ylabel("Average Price (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.yscale("log")
plt.show()
```

```python
# Line chart: Number of artworks sold per year
yearly_sales = art["sold_year"].value_counts().sort_index()

plt.plot(yearly_sales.index, yearly_sales.values, marker='o', linestyle='-', color='navy')
plt.title("Artworks Sold Over the Years")
plt.xlabel("Sold Year")
plt.ylabel("Number of Artworks")
plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
# Number of artworks sold per auction house
auction_counts = art["auction_house"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    auction_counts,
    labels=auction_counts.index,
    autopct="%1.1f%%",
    startangle=140,
    colors=sns.color_palette("pastel")
)
plt.title("Artwork Distribution by Auction House")
plt.tight_layout()
plt.show()

```

```python
# Plot histogram showing distribution of artwork widths
sns.histplot(art["width_cm"], bins=40, color='lightblue')
plt.title("Distribution of Artwork Width")
plt.xlabel("Width (cm)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
